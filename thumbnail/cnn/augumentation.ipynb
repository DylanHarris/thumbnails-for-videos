```python
import glob, os, os.path
import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import ImageFilter
dir_ = './ppt/'
dest_dir = './ppt/'

count = 10

filelist = glob.glob(os.path.join(dir_, "*.jpg"))
for f in filelist:

    image = Image.open(f)
    
    img2 = image.point(lambda p: p * 0.25)
    img2.save(dest_dir + 'imaged' + str(count) + '.jpg')


    img3 = image.filter(ImageFilter.GaussianBlur(10))
    img3.save(dest_dir + 'imagegb' + str(count) + '.jpg')


    img6 = image.filter(ImageFilter.UnsharpMask(radius=200, percent=500, threshold=1))
    img6.save(dest_dir + 'imageshar' + str(count) + '.jpg')
    
    count += 1
```
