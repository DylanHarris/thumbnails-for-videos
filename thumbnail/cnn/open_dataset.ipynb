```python
count = 1
label_dict = {}
with open('./open/2017_11/train/annotations-human.csv', 'r') as f:
    for line in f:
        #print line
        if count:
            count = 0
            continue
        data = line.strip().split(",")
        label = data[-2]
        confi = int(data[-1])
        if confi == 1:
            if label not in label_dict:
                label_dict[label] = []
            label_dict[label].append(data[0])
```

```python
print len(label_dict)
final_dict = {}
import copy
count = 0
for i in label_dict:
    if len(label_dict[i]) > 5:
        count += 1
        final_dict[i] = copy.deepcopy(label_dict[i][:2])
```

```python
flag = 1
image = {}
with open('./open/2017_11/train/images.csv', 'r') as f:
    for line in f:
        if flag:
            flag = 0
            continue
        data = line.strip().split(",")
        try:
            image[data[0]] = data[-1]
        except:
            pass
```
